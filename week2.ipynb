# 證券交易所 TWSE

```python
import requests
import pandas as pd
from datetime import datetime

today = datetime.now().strftime("%Y%m%d")
stock_id = "2330"
url = f"https://www.twse.com.tw/rwd/zh/afterTrading/STOCK_DAY?date={today}&stockNo={stock_id}&response=json"
response = requests.get(url)
data = response.json()

df = pd.DataFrame(data["data"], columns=data["fields"])
df
```

# Yahoo

```python
import bs4

stock_id = "2330"
url = f"https://tw.stock.yahoo.com/quote/{stock_id}.TW"
response = requests.get(url)
soup = bs4.BeautifulSoup(response.text, "html.parser")

price = soup.find_all("span", class_="Fw(600)")
columns = ["股票代號", "成交", "開盤", "最高", "最低", "均價", "成交金額(億)", "昨收", "漲跌幅", "漲跌", "總量", "昨量", "振幅"]

price_text = [p.text for p in price[:len(columns) - 1]]
data = [stock_id] + price_text

df = pd.DataFrame([data], columns=columns)
df
```
